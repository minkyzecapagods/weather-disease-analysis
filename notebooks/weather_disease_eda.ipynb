Este projeto segue as fases do CRISP-DM, com foco intenso nas etapas iniciais (típicas de uma EDA):

| Fase | Ação |

| 1. Compreensão do negócio | Definir perguntas de pesquisa (impacto clima ↔ saúde) |

## Objetivo (editar)

Realizar uma análise exploratória  para identificar padrões, tendências e correlações entre:

- Incidência de arboviroses (dengue, chikungunya, zika)
- Incidência de doenças respiratórias (SRAG, COVID-19, influenza)
- Variáveis climáticas: temperatura média, precipitação acumulada e estação ou época do ano
