```python
import requests
import zipfile
import io
import pandas as pd

# Anos que pra baixar
anos = ['24', '25']

# Dicionário para guardar cada DataFrame (chave = ano)
dfs = {}

for ano in anos:
    url = f"https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/SINAN/Dengue/csv/DENGBR{ano}.csv.zip"
    print(f"Baixando DENGBR{ano}...")

    # 1. Faz o download do ZIP em memória (sem salvar no disco)
    response = requests.get(url, stream=True)
    if response.status_code != 200:
        print(f"Erro ao baixar {ano}: status {response.status_code}")
        continue

    # 2. Abre o ZIP e extrai o CSV
    with zipfile.ZipFile(io.BytesIO(response.content)) as z:
        # Normalmente o CSV tem o mesmo nome do ZIP, ex: DENGBR22.csv
        nome_csv = f"DENGBR{ano}.csv"
        with z.open(nome_csv) as f:
            # 3. Carrega no Pandas
            df = pd.read_csv(
                f,
                sep=',',
                low_memory=False,
                dtype=str,
                encoding='latin1'
            )
            # Adiciona coluna de ano da base, se quiser
            df['ARQUIVO_ANO'] = f"20{ano}"
            dfs[ano] = df
            print(f"DENGBR{ano}: {df.shape[0]} linhas, {df.shape[1]} colunas")
```

```python
#CÉLULA 1 — DOWNLOAD VIA API INFODENGUE
# Baixa dados de dengue de Natal/RN (geocódigo 2408102) para 2019–2026.
# Estratégia: tenta CSV primeiro; se falhar, tenta JSON; por fim, fallback
# manual via requests com tratamento de timeout e erros HTTP.

BASE_URL = 'https://info.dengue.mat.br/api/alertcity'

# ── Parâmetros da consulta ────────────────────────────────────────────────────
PARAMETROS = {
    'geocode' : 2408102,   # Natal/RN
    'disease' : 'dengue',
    'ew_start': 1,
    'ew_end'  : 53,
    'ey_start': 2019,
    'ey_end'  : 2026,
}

# ── Função robusta de download ────────────────────────────────────────────────
def baixar_dados_infodengue(params: dict, formato: str = 'csv',
                             timeout: int = 60) -> pd.DataFrame | None:
    """
    Consulta a API InfoDengue e retorna um DataFrame.

    Parameters
    ----------
    params  : dict  – parâmetros da URL (geocode, disease, etc.)
    formato : str   – 'csv' ou 'json'
    timeout : int   – tempo máximo de espera em segundos

    Returns
    -------
    pd.DataFrame ou None se a consulta falhar.
    """
    url_params = '&'.join(f'{k}={v}' for k, v in {**params, 'format': formato}.items())
    url = f'{BASE_URL}?{url_params}'
    print(f' Consultando ({formato.upper()}): {url}')

    try:
        resp = requests.get(url, timeout=timeout)
        resp.raise_for_status()

        if formato == 'csv':
            # Tenta múltiplos separadores (vírgula e ponto-e-vírgula)
            content = resp.text
            sep = ';' if content.count(';') > content.count(',') else ','
            df = pd.read_csv(StringIO(content), sep=sep,
                             encoding='utf-8', on_bad_lines='skip')
        else:
            data = resp.json()
            # A API pode retornar lista ou dict
            if isinstance(data, list):
                df = pd.DataFrame(data)
            elif isinstance(data, dict):
                # Às vezes retorna {col: {idx: val}}
                df = pd.DataFrame.from_dict(data)
            else:
                print(f'  Formato JSON inesperado: {type(data)}')
                return None

        # Remove linhas completamente nulas
        df = df.dropna(how='all')
        print(f' {len(df):,} linhas × {df.shape[1]} colunas baixadas.')
        return df

    except requests.exceptions.Timeout:
        print(f'  ⏱  Timeout após {timeout}s.')
    except requests.exceptions.HTTPError as e:
        print(f'  ✗  Erro HTTP: {e}')
    except Exception as e:
        print(f'  ✗  Erro inesperado: {e}')
    return None


# ── Execução com estratégia de fallback ──────────────────────────────────────
print('=' * 65)
print('INICIANDO DOWNLOAD — DENGUE NATAL/RN (2019–2026)')
print('=' * 65)

df_raw = None

# Tentativa 1: CSV
df_raw = baixar_dados_infodengue(PARAMETROS, formato='csv')

# Tentativa 2: JSON (se CSV falhar)
if df_raw is None or df_raw.empty:
    print('\n  ↩  Tentando formato JSON como fallback...')
    df_raw = baixar_dados_infodengue(PARAMETROS, formato='json')

# Tentativa 3: Download por partes (ano a ano) se a consulta total falhar
if df_raw is None or df_raw.empty:
    print('\n  ↩  Tentando download por ano (fallback final)...')
    partes = []
    for ano in range(PARAMETROS['ey_start'], PARAMETROS['ey_end'] + 1):
        p = {**PARAMETROS, 'ey_start': ano, 'ey_end': ano}
        df_part = baixar_dados_infodengue(p, formato='json', timeout=30)
        if df_part is not None and not df_part.empty:
            partes.append(df_part)
    if partes:
        df_raw = pd.concat(partes, ignore_index=True)
        print(f' Consolidado: {len(df_raw):,} linhas')

# ── Resultado ─────────────────────────────────────────────────────────────────
if df_raw is not None and not df_raw.empty:
    print('\n' + '=' * 65)
    print(' DOWNLOAD CONCLUÍDO')
    print('=' * 65)
    print(f' Shape bruto : {df_raw.shape}')
    print(f' Colunas     : {list(df_raw.columns)}')
    print()
    display(df_raw.head(3))
else:
    print('\n FALHA NO DOWNLOAD. Verifique a conexão ou a disponibilidade da API.')
    # Cria DataFrame vazio com as colunas esperadas para permitir execução das
    # células seguintes mesmo sem conexão (útil para demonstração offline)
    df_raw = pd.DataFrame(columns=[
        'SE','data_iniSE','casos_est','casos_est_min','casos_est_max',
        'casos','p_rt1','p_inc100k','Rt','pop','tempmin','tempmed','tempmax',
        'umidmin','umidmed','umidmax','receptivo','transmissao','nivel_inc',
        'nivel','id'
    ])
    print('  ⚠  DataFrame vazio criado. Execute novamente quando a API estiver disponível.')
```
