```python
import requests
import zipfile
import io
import pandas as pd
from pathlib import Path
```

```python
# ---------- CONFIGURAÇÕES ----------
DOENCAS = {
    'DENG': 'Dengue',
    'CHIK': 'Chikungunya',
    'ZIKA': 'Zikavirus'
}

ANOS = list(range(2016, 2026))  # 2016 a 2025
ANOS_STR = [str(a)[-2:] for a in ANOS]  # ['16','17',...,'25']

# Códigos IBGE das cidades (Natal, Santa Cruz, Caicó, Mossoró, ...)
CODIGOS_ALVO = ['240810', '241200', '240200', '240800']

# Pasta para salvar os parquets (cache)
PASTA_CACHE = Path('./cache_arboviroses')
PASTA_CACHE.mkdir(exist_ok=True)

# URL padrão
URL_BASE = "https://s3.sa-east-1.amazonaws.com/ckan.saude.gov.br/SINAN/{doenca_nome}/csv/{doenca_sigla}BR{ano_str}.csv.zip"
```

```python
def processar_arquivo(doenca_sigla, doenca_nome, ano_int):
    """Baixa, filtra e agrega os casos por semana/município.
       Salva o agregado em Parquet e o retorna."""

    ano_str = str(ano_int)[-2:]
    url = URL_BASE.format(doenca_nome=doenca_nome, doenca_sigla=doenca_sigla, ano_str=ano_str)
    nome_cache = PASTA_CACHE / f"{doenca_sigla}_{ano_int}.parquet"

    # Se já existe cache, apenas carrega
    if nome_cache.exists():
        print(f"  Cache encontrado para {doenca_nome} {ano_int}. Carregando...")
        return pd.read_parquet(nome_cache)

    print(f"  Baixando {doenca_nome} {ano_int}...")
    response = requests.get(url, stream=True)
    if response.status_code != 200:
        print(f"    Erro HTTP {response.status_code}. Pulando.")
        return None

    chunks_filtrados = []

    try:
        with zipfile.ZipFile(io.BytesIO(response.content)) as z:
            nome_csv = f"{doenca_sigla}BR{ano_str}.csv"
            with z.open(nome_csv) as f:
                reader = pd.read_csv(
                    f,
                    sep=',',
                    chunksize=100000,
                    dtype=str,
                    encoding='latin1',
                    low_memory=False
                )

                for chunk in reader:
                    # Filtra pelo local provável de infecção (COMUNINF)
                    if 'COMUNINF' in chunk.columns:
                        chunk = chunk[chunk['COMUNINF'].isin(CODIGOS_ALVO)]
                    else:
                        print("    Coluna COMUNINF não encontrada!")
                        break

                    if not chunk.empty:
                        chunks_filtrados.append(chunk)

    except Exception as e:
        print(f"    Erro ao ler o arquivo: {e}")
        return None

    if not chunks_filtrados:
        print(f"    Nenhum caso encontrado para as cidades em {ano_int}.")
        return None

    # Concatena todos os chunks
    df = pd.concat(chunks_filtrados, ignore_index=True)
    print(f"    Total de registros filtrados: {df.shape[0]}")

    # Processamento: mantém apenas SEM_PRI e COMUNINF, agrega
    cols_necessarias = ['COMUNINF', 'SEM_PRI']
    df = df[cols_necessarias].copy()

    # Remove linhas sem semana epidemiológica (caso raro)
    df = df.dropna(subset=['SEM_PRI'])

    # Agregação
    agregado = df.groupby(['COMUNINF', 'SEM_PRI']).size().reset_index(name='CASOS')
    agregado = agregado.rename(columns={
        'COMUNINF': 'MUNICIPIO',
        'SEM_PRI': 'SEMANA_EPI'
    })
    agregado['DOENCA'] = doenca_nome.upper()
    agregado['ANO'] = ano_int

    # Remove semanas epidemiológicas com ano anterior a 2016 (ruído)
    agregado = agregado[agregado['SEMANA_EPI'].str[:4].astype(int) >= 2016]

    # Salva cache
    agregado.to_parquet(nome_cache)
    print(f"    Agregado salvo em {nome_cache}")

    return agregado
```

```python
# Lista para guardar todos os DataFrames agregados
todos_agregados = []

for sigla, nome_doenca in DOENCAS.items():
    print(f"\n===== {nome_doenca.upper()} =====")
    for ano in ANOS:
        agregado = processar_arquivo(sigla, nome_doenca, ano)
        if agregado is not None:
            todos_agregados.append(agregado)

# Concatena tudo em um único DataFrame
df_arbovirores = pd.concat(todos_agregados, ignore_index=True)

print("\n===== Processamento concluído! =====")
print(f"Shape final: {df_arbovirores.shape}")
display(df_arbovirores.head())
```
